```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import json
import csv

url = 'dataset_real_stars.csv'

mapping = {
    'beer_and_wine': 1,
    'full_bar':2,
    'none': -1,
    'free': 1,
    'paid': 2,
    'outdoor': 1,
    'casual': 1,
    'dressy': 2,
    'formal': 4,
    'average': 1,
    'loud': 2,
    'quiet': 3,
    'very_loud': 4,
    'yes_corkage': 1
}

states = {
    'AK': 1, 'AL': 2, 'AR': 3, 'AZ': 4, 'CA': 5, 'CO': 6, 'CT': 7, 'DE': 8, 'FL': 9, 'GA': 10, 'HI': 11, 'IA': 12,
    'ID': 13, 'IL': 14, 'IN': 15, 'KS': 16, 'KY': 17, 'LA': 18, 'MA': 19, 'MD': 20, 'ME': 21, 'MI': 22, 'MN': 23, 
    'MO': 24, 'MS': 25, 'MT': 26, 'NC': 27, 'ND': 28, 'NE': 29, 'NH': 30, 'NJ': 31, 'NM': 32, 'NV': 33, 'NY': 34,
    'OH': 35, 'OK': 36, 'OR': 37, 'PA': 38, 'RI': 39, 'SC': 40, 'SD': 41, 'TN': 42, 'TX': 43, 'UT': 44, 'VA': 45, 
    'VT': 46, 'WA': 47, 'WI': 48, 'WV': 49, 'WY': 50, 'AB': 51, 'QC': 52, 'ON': 53, 'NYK': 54, 'VS': 55, 'NI': 56,
    'BY': 57, 'O': 58, 'C': 59, 'SP': 60, 'BC': 61
}

df = pd.read_csv(url,low_memory=False)

df = df.replace({'attributes__Alcohol': mapping,
                 'state': states,
                 'attributes__WiFi': mapping,
                 'attributes__Smoking': mapping,
                 'attributes__RestaurantsAttire': mapping,
                 'attributes__NoiseLevel': mapping,
                 'attributes__BYOBCorkage': mapping
                })

df = pd.DataFrame(df).fillna(0)
df.head()
```

```python
features = ['attributes__Alcohol', 'attributes__Ambience__classy', 'attributes__Ambience__hipster']

features = df.keys().drop({'business_id', 'state', 'attributes__DietaryRestrictions', 'business_id.1'})

x = df.loc[:, features]

y = df.loc[:,['avg(stars)']].values

pd.DataFrame(data = x, columns = features).head()

```

```python
pca = PCA(n_components=2)

transformed = pca.fit_transform(x)

principalDf = pd.DataFrame(data = transformed
             , columns = ['principal component 1', 'principal component 2'])

principalDf.head(5)
```

```python

df[['stars']].head()
```

```python
finalDf = pd.concat([principalDf, df[['stars']]], axis = 1)
finalDf.head(5)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = [1, 2, 3, 4, 5]
colors = ['r', 'g', 'b', 'y', 'm']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['stars'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
